```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv(r"C:\Coding\Machine_Learning\Machine Learning Coding\Polynomial Regression\Practice Problem\Closed Form\Dataset\Ice_cream selling data.csv")
df.head()
```

```python
x = df['Temperature (°C)']
y = df['Ice Cream Sales (units)']
```

```python
degree = 2
X_Raw = np.column_stack([x**i for i in range (1,degree + 1)])
X_Scaled = (X_Raw - np.mean(X_Raw)) / np.std(X_Raw)
Y_Scaled = np.array((y - np.mean(y)) / np.std(y))
Y_Scaled = Y_Scaled.reshape(-1,1)
```

```python
X = np.column_stack([np.ones(len(X_Scaled)), X_Scaled])
```

```python
iterations = 200

ALPHA = 0.017
w = np.zeros(X.shape[1]).reshape(-1,1)
cost_history = []

for i in range (iterations):
    Y_Pred = X @ w

    cost = (1/(2*len(df)) * np.sum((Y_Pred-Y_Scaled)**2))
    cost_history.append(cost)

    gradient = X.T @ (Y_Pred - Y_Scaled)
    w = w - ALPHA * gradient

        
        
print (w)
print(cost)
```
